# YOLOv3 Test using Darknet Framework and pre-trained weights

```python
import cv2
import matplotlib.pyplot as plt

from utils import *
from darknet import Darknet
```

```python
cfg_file = './cfg/yolov3.cfg'
weight_file = './weights/yolov3.weights'

namesfile = 'data/coco.names'

# Load network from config and the weights from weight file
m = Darknet(cfg_file)
m.load_weights(weight_file)

# Load classes from COCO dataset
class_names = load_class_names(namesfile)
```

```python
# Load an image of my dogs
img = cv2.imread('./images/2dogs.jpg')
```

```python
#Decresaing non-max suppression allows more boxes to be found and more potential localizations 
nms_thresh = 0.6 
#Increasing Intersection over Union can result in multiple boxes being output since 
iou_thresh = 0.4
```

```python
# Set the default figure size
plt.rcParams['figure.figsize'] = [24.0, 14.0]

# Convert the image to RGB
original_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize the image to match first network layer  
resized_image = cv2.resize(original_image, (m.width, m.height))

# Detect objects in the image using function in utils file 
# Uses Pytorch backend 
boxes = detect_objects(m, resized_image, iou_thresh, nms_thresh)

# Print the objects found and the confidence level
print_objects(boxes, class_names)

#Plot the image with bounding boxes and corresponding object class labels
plot_boxes(original_image, boxes, class_names, plot_labels = True)
```
